## Season ,Payment choice and Number of transaction

```python
seasons = ["winter","spring","summer","autumn"]
payment_choices = ["cash","credit_card"]
N = 15000
```

```python
data = {
    "name": [fake.name() for _ in range(1001)],
    "email": [fake.email() for _ in range(1001)],
    "phone": [fake.phone_number() for _ in range(1001)],
    "adress": [fake.address() for _ in range(1001)],
    "birthday": [fake.date_of_birth(minimum_age=18, maximum_age=80) for _ in range(1001)],
}
df2 = pd.DataFrame(data)
```

```python
bonus_cards = [fake.unique.bothify(text="BONUS-????-####") for _ in range(1000)]

bonus_usage = {card: 0 for card in bonus_cards}

records1 = []
records2 = []

for _ in range(N):
    tid = fake.uuid4()
    date = fake.date_time_between(start_date="-1y", end_date="now")
    season = seasons[(date.month - 1) // 3]
    pay = random.choice(payment_choices)

    if random.random() < 0.8:
        avail = [c for c, cnt in bonus_usage.items() if cnt < 10]
        if avail:
            bonus = random.choice(avail)
            bonus_usage[bonus] += 1
        else:
            bonus = None
    else:
        bonus = None

    # məhsul və qiymətlər
    count = random.randint(1, 5)
    items = random.sample(products, count)
    subtotal = sum(i["price"] for i in items)
    tax = round(subtotal * 0.1, 2)
    total = round(subtotal + tax, 2)
    review = random.randint(1, 5)

    records1.append({
        "transaction_id": tid,
        "bonus_card_number": bonus,
        "total_price": total,
        "date": date,
        "season": season,
        "customer_review": review,
        "payment_method": pay
    })
    records2.append({
        "transaction_id": tid,
        "total_price": total,
        "purchased_products": [i["name"] for i in items]
    })

df1 = pd.DataFrame(records1)
df3 = pd.DataFrame(records2)
```
